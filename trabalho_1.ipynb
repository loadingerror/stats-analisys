date: Data em que ocorreu o registro.
publisherId: Identificador da loja que solicitou o anúncio.

adType: Tipo de anúncio exibido.
searchTerm: Termo de pesquisa associado à solicitação do anúncio.
zipcode: Código postal do cliente logado.
totalAdsRequestedCount: Número total de anúncios solicitados.
totalAdsReturnedCount: Número total de anúncios retornados.
totalRequests: Total de solicitações realizadas.

```python
import pandas as pd
import pickle
```

```python
# Abrir o arquivo em modo de leitura binária ('rb' para read binary)
with open('ads-fillrate-daily-search-summary.pkl', 'rb') as file:
    # Carregue o objeto Pickle
    df = pickle.load(file)
```

```python
# Imprime as 5 primeiras linhas do Dataframe
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Converta a coluna 'date' para o tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Crie uma nova coluna 'year_month' contendo o ano e o mês
df['year_month'] = df['date'].dt.strftime('%Y-%m')
```

```python
# Novo DataFrame com apenas as colunas desejadas retirei id, page, adType
nw_df = df[['date', 'year_month', 'publisherId','searchTerm', 'zipcode', 'totalAdsRequestedCount', 'totalAdsReturnedCount', 'totalRequests']].copy()
```

```python
nw_df.head()
```

```python
pivottable_1 = pd.pivot_table(df, index='publisherId', 
                                  values=['totalAdsRequestedCount', 'totalAdsReturnedCount', 'totalRequests'], 
                                  aggfunc='sum')
```

```python
# Exiba a tabela dinâmica
print(pivottable_1)
```

```python
# Aplicar o filtro para incluir apenas as linhas onde publisherId é "netshoes"
df_filtrado = df[df['publisherId'] == 'netshoes']

# Criar a tabela dinâmica usando pivot_table()
pivottable_2 = pd.pivot_table(df_filtrado, index=['publisherId', 'searchTerm'], 
                                  values=['totalAdsRequestedCount', 'totalAdsReturnedCount', 'totalRequests'], 
                                  aggfunc='sum')
```

```python
print(pivottable_2)
```
